## 📝 Task 2 — Grouped Bar Chart with Matplotlib

### Background
A **grouped bar chart** places multiple bars side by side for the same category, making it easy to compare values within each group.

### 🎯 Your Task
The data below shows **exam scores** for three subjects across two classes.

| Subject | Class A | Class B |
|---------|---------|---------|
| Math | 72 | 68 |
| Science | 85 | 80 |
| English | 78 | 74 |
| History | 65 | 70 |

Create a **grouped bar chart** with:
- Two bars (Class A and Class B) side by side for each subject
- Different colors for each class
- Value labels on top of each bar
- A legend, axis labels, and a title

> 💡 **Hint:** Use `np.arange(len(subjects))` for x-positions, then offset each group with `x - 0.2` and `x + 0.2` and `width=0.4`.  
> To add labels: loop over bars using `bar.get_height()` and `plt.text()`.


````python
import numpy as np

subjects = ['Math', 'Science', 'English', 'History']
class_a  = [72, 85, 78, 65]
class_b  = [68, 80, 74, 70]

# YOUR CODE HERE
# Create x positions using np.arange
x = np.arange(len(subjects))
width = 0.4

# Plot Class A bars (offset left)
bar1 = plt.bar(x - width/2, class_a, width, label='Class A', color='blue')

# Plot Class B bars (offset right)
bar2 = plt.bar(x + width/2, class_b, width, label='Class B', color='green')

# Add value labels above each bar
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height, f'{height}', ha='center', va='bottom')
# Add legend
plt.legend()
# Add title, axis labels, legend, and x-tick labels
plt.title("Exam Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.xticks(x, subjects)
plt.show()
````

## 📝 Task 7 — Interactive Line Plot with Bokeh

### Background
Bokeh creates **interactive, browser-rendered** plots. The `HoverTool` lets users see exact data values when they hover over points.

### 🎯 Your Task
Create an **interactive Bokeh line plot** showing website traffic over 7 days:

```
days    = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]
```

Requirements:
- Line width = 2, color of your choice
- Circle markers on each data point (size = 8)
- A `HoverTool` that shows the day and visitor count on hover
- Proper title and axis labels

> 💡 **Hint:** To use HoverTool with plain lists, convert data to a `ColumnDataSource`.  
> `source = ColumnDataSource(data=dict(days=days, visitors=visitors))`  
> Then use `p.line(x='days', y='visitors', source=source)` and  
> `HoverTool(tooltips=[("Day", "@days"), ("Visitors", "@visitors")])`.


````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# YOUR CODE HERE
# 1. Create a ColumnDataSource
source = ColumnDataSource(data=dict(days=days, visitors=visitors))

# 2. Create a Bokeh figure with title and axis labels
p = figure(x_range=days, height=350, title="Website Traffic Over 7 Days",
           x_axis_label="Day", y_axis_label="Visitors")

# 3. Add a line renderer
p.line(x='days', y='visitors', source=source, line_width=2, color='blue')

# 4. Add circle markers
p.circle(x='days', y='visitors', source=source, size=8, color='blue', fill_alpha=0.5)

# 5. Add a HoverTool
hover = HoverTool(tooltips=[
    ("Day", "@days"),
    ("Visitors", "@visitors"),
])
p.add_tools(hover)

# 6. Show the plot
show(p)
````

# ── Chart 4 (Optional) ──────────────────────────────────────────────────

# Why it suits this data: To compare two quantitative variables ('Returns' and 'Orders') across a categorical variable ('Day'). The side-by-side bars make it easy to see the absolute difference and relationship between returns and orders for each day of the week.

````python
import plotly.graph_objects as go
import pandas as pd

data = {
    'Day':        ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Orders':     [45,   62,   58,   70,   95,   110,  80 ],
    'Revenue':    [900,  1240, 1160, 1400, 1900, 2200, 1600],
    'Returns':    [2,    3,    5,    4,    6,    8,    5   ],
    'Category':   ['Electronics','Clothing','Books','Electronics','Clothing','Books','Electronics']
}
df = pd.DataFrame(data)

# Create traces for Orders and Returns
fig = go.Figure(data=[
    go.Bar(name='Orders', x=df['Day'], y=df['Orders']),
    go.Bar(name='Returns', x=df['Day'], y=df['Returns'])
])

# Update layout for grouped bar chart
fig.update_layout(
    barmode='group',
    title_text='Orders vs Returns by Day',
    xaxis_title='Day',
    yaxis_title='Count'
)

fig.show()
````

````python
# Q1: What is the main advantage of using Seaborn over raw Matplotlib for statistical plots?
# A1: Seaborn provides a higher-level interface for drawing attractive and informative statistical graphics.
# It integrates well with pandas DataFrames, automatically handles many common statistical transformations,
# and offers a wide range of specialized plot types (e.g., displot, heatmap, lmplot) with fewer lines of code
# compared to Matplotlib, which requires more explicit control over plot elements.

# Q2: When would you choose Bokeh over Plotly, and vice versa?
# A2: Choose Bokeh when you need highly customizable, interactive plots for web applications, especially
# for streaming data or complex dashboards where fine-grained control over interactivity and layout is crucial.
# Bokeh generates standalone HTML or can be embedded into web applications.
# Choose Plotly (especially Plotly Express) when you need quick, high-quality, interactive statistical plots
# with less code. Plotly is excellent for Exploratory Data Analysis (EDA) and reporting, offering a wide range
# of chart types and easy animation/filtering capabilities, often with better out-of-the-box styling and a more
# intuitive API for common tasks. Plotly's `go.Figure` offers similar low-level control to Bokeh.

# Q3: In Task 3, how does changing the number of bins affect your interpretation of the distribution?
# A3: Changing the number of bins in a histogram affects the granularity of the distribution. Too few bins can
# oversimplify the distribution, hiding important features like multiple peaks or gaps. Too many bins can make
# the distribution appear noisy and choppy, making it difficult to discern overall patterns. The goal is to
# choose a number of bins that reveals the underlying shape of the data without being too coarse or too fine.

# Q4: Why is a pie chart generally not recommended when you have more than 5–6 categories?
# A4: Pie charts become difficult to read and interpret with too many categories because the slices become
# too thin, making it hard to accurately compare their relative sizes. It's challenging for the human eye
# to compare angles or areas effectively when there are many small segments. Bar charts are generally preferred
# for comparing multiple categories as they allow for easier visual comparison of lengths.

# Q5: What does a violin plot show that a box plot does not?
# A5: A violin plot shows the full probability density of the data at different values (like a smoothed histogram
# rotated on its side) for each category, in addition to the statistical summaries provided by a box plot (median,
# quartiles, and outliers). This means a violin plot can reveal the presence of multiple modes (peaks) in the
# distribution, skewness, and the density of data points across the range, which a box plot cannot show.
````
